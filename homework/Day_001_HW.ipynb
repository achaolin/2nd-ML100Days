HW 1-1：
資料選擇 Air Quality Taiwan 2014-2018
1. 資料為何重要
   空氣品質的好壞與身體健康息息相關，若能預測隔日的空氣品質，便能提前作準備以預防吸入過多不好的空氣。
2. 資料來源
   台灣行政院環保署空氣品質監測網
3. 資料型態
   大部分的欄位皆為數值型資料，包含SO2、CO、PM2.5等地含量，但也有少部分如空氣品質監測站的位置、國家代碼/名稱等為類別型資料
4. 問題如何評估
   空氣品質會以空氣品質指標(AQI)來評估，AQI可依據監測站測出空氣中O3、PM2.5、PM10、CO、SO2、NO2濃度來評估對人體健康程度的影響，先將其轉換成副指標值，再依據每日的最高值為當日AQI。
   因AQI為一個數值型的指標，因此若要評估預測的AQI是否準確，可以均方誤差去評估預測值與實際值的差異。

HW 1-2：
1. 核心問題
   假設業績與載客趟數呈正相關，趟數越多代表業績越好，因此，提升業績的方式為增加載客趟數。
2. 資料來源
   使用每台車上跳錶的裝置來統計載客趟數，每結算一次代表一趟，並且同時記錄距離、所花的時間等。每天每台車上的數據回傳到雲端資料庫中，再重資料庫中下載資料。
   除了距離、趟數、時間可能影響業績外，顧客滿意度、回搭率等也都是影響業績的一項重要因子，因此還需建立一個顧客評比及推薦系統，讓使用者填寫，並於資料庫中統一下載。另外天氣、地區等外在因素也可能會影響顧客願意乘車的意願，因此也需一併記錄於資料庫中。
3. 資料型態
   載客距離、時間、趟數等皆為數值型資料；顧客滿意度、推薦度等為類別型資料。
4. 如何評估
   先將滿意度等類別型資料依照最滿意、滿意、普通、不滿意、很不滿意五個等級分別轉換成5、4、3、2、1五個分數，接著再利用其餘資料建立一個回歸模型。輸出變數為載客趟數，輸入變數則為其餘蒐集到的資料。建立模型後，便可預測每個司機今天可能的業績。有了預測值與實際值後，再利用均方誤差來評估模型的準確度。
